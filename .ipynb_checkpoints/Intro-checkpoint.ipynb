#Convert 1, 10, 20, 30, 100 from your base to base10.

```python
print(int('1', 7))     # Base 7 -> Base 10 = 1
print(int('10', 7))    # 1*7^1 + 0 = 7
print(int('20', 7))    # 2*7^1 + 0 = 14
print(int('30', 7))    # 3*7^1 + 0 = 21
print(int('100', 7))   # 1*7^2 + 0 + 0 = 49
```

#Write a strategy to convert a number in your base to base 10 in plain english or pseudo code or python


```python
def extract(mult, result, num):
    rem = num % 10
    result = result + rem*mult
    num = num // 10
    mult = mult * 7
    return mult, result, num
```

```python
mult, result, num = 1, 0, 4123
```

```python
mult, result, num = extract(mult, result, num)
print(mult, result, num)
```

```python
#Write a strategy to convert a number in base N to base M in plain english or pseudo code or python
```

```python
def convert_base(num_str, from_base, to_base):
    # Step 1: Convert from base-N to base-10
    base10 = 0
    power = 0
    for digit in reversed(num_str):
        base10 += int(digit) * (from_base ** power)
        power += 1

    # Step 2: Convert from base-10 to base-M
    if base10 == 0:
        return "0"

    result = []
    while base10 > 0:
        remainder = base10 % to_base
        result.append(str(remainder))
        base10 //= to_base

    result.reverse()
    return ''.join(result)

# Example: Convert '203' from base 4 to base 5
print(convert_base("203", 4, 5))  # Output: "120"
```
